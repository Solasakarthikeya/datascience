```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Universities_Clustering.csv')
df
```

```python
len(df)
```

```python
df.iloc[:,:]
```

```python
df.iloc[0:len(df)-1,:]
```

```python
plt.hist(df['SAT'])
plt.xlabel('SAT')
plt.ylabel('Frequency')
plt.title('Histogram of SAT')
plt.show()
```

```python
plt.hist(df['Accept'])
plt.xlabel('Accept')
plt.ylabel('Frequency')
plt.title('Histogram of Accept')
plt.show()
```

```python
sns.histplot(df['Accept'])
plt.xlabel('ACC')
plt.ylabel('Frequency')
plt.title('Histogram of ACC')
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.distplot(df['Accept'])
plt.xlabel('ACC')
plt.ylabel('Frequency')
plt.title('Histogram of ACC')
plt.show()
```

```python
plt.plot(df['Accept'],df['SAT'],'o') #bo : blue colour circles
plt.xlabel('ACC')
plt.ylabel('SAT')
plt.title('Line Plot of ACC and SAT')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df.iloc[:,0:4])
```

```python
plt.figure(figsize=(10,10))
plt.boxplot(df['Accept'],vert=False)
plt.xlabel('ACC')
plt.ylabel('Frequency')
plt.title('Box Plot of ACC')
plt.show()
```
